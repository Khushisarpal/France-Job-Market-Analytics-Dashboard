```python
from jobspy import scrape_jobs
import pandas as pd
from datetime import datetime

jobs = scrape_jobs(
    site_name=["indeed"],
    search_term="data analyst",
    location="France",
    results_wanted=100,
    hours_old=168,        # last 7 days only
    country_indeed="France",
    language="fr",
    linkedin_fetch_description=False
)

# clean up
jobs = jobs[[
    "title", "company", "location", "job_type",
    "min_amount", "max_amount", "description",
    "job_url", "date_posted"
]]

jobs.to_csv("jobs_france.csv", index=False, encoding="utf-8-sig")
print(f"Scraped {len(jobs)} jobs at {datetime.now()}")
```
